## Introduction to line plots

- What are line plots?
    - Two types of relational plots: scaer plots and line plots

- Scatter plots
    - Each plot point is an independent observation

- Line plots
    - Each plot point represents the same "thing" , typically tracked over time

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Create line plot
sns.relplot(x="model_year",y="mpg",data=mpg,kind="line")


# Show plot
plt.show()

#The shaded region represents a confidence interval for the mean, not the distribution of the observations
```

### Multiple observations per x-value
- Shaded region is the confidence interval
    - Assumes dataset is a random sample
    - 95% condent that the mean is within this interval
    - Indicates uncertainty in our estimate

```python
# Make the shaded area show the standard deviation
sns.relplot(x="model_year", y="mpg",data=mpg, kind="line",ci="sd")

# Show plot
plt.show()
#this plot shows us the distribution of miles per gallon for all the cars in each year.
```

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Create line plot of model year vs. horsepower
sns.relplot(x="model_year",y="horsepower",kind="line",data=mpg,ci=None)

# Show plot
plt.show()
```

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Change to create subgroups for country of origin
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            style="origin",
            hue="origin",
            dashes=False,  # Getting solid line instead of dashes
            markers=True, # Getting markers
            ci=None)

# Show plot
plt.show()

## we can see that this downward trend in horsepower was more pronounced among cars from the USA.
```

## Creating a box plot
- What is a box plot?
- Shows the distribution of quantitative data
- See median, spread, skewness, and outliers
- Facilitates comparisons between groups

```python
tips
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g = sns.catplot(x="day",y="total_bill",data=tips,kind="box")
#g = sns.catplot(x="day",y="total_bill",data=tips,kind="box",whis=[0,100])

plt.xlabel("Day of the Week")
plt.ylabel("Total Bill(in dollars)")
plt.show()

# Box ?
# Middle line in box ? = The line in the middle of each box represents the median.

# Upper and lower T shape = These called whiskers if we use whis=[0,100] then our whiskers will include all outliers
    # Changing the whiskers using `whis`
    # By default, the whiskers extend to 1.5 * the interquartile range
    # Make them extend to 2.0 * IQR: whis=2.0
    # Show the 5th and 95th percentiles: whis=[5, 95]
    # Show min and max values: whis=[0, 100]

# IQR
# https://www.statology.org/how-to-interpret-interquartile-range/
# https://www.statology.org/interquartile-range-calculator/    
# https://statisticsbyjim.com/basics/interquartile-range/#:~:text=What%20is%20the%20Interquartile%20Range,most%20of%20your%20values%20lie.

# Dots outside = Outliers
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g = sns.catplot(x="day",y="total_bill",data=tips,kind="box",whis=[0,100])

plt.xlabel("Day of the Week")
plt.ylabel("Total Bill(in dollars)")
plt.show()
```

## Point plots
- What are point plots?
- Points show mean of quantitative variable
- Vertical lines show 95% confidence intervals

- Point plots vs. line plots
- Both show:
    - Mean of quantitative variable
    - 95% condence intervals for the mean
- Differences:
    - Line plot has quantitative variable (usually time) on x-axis
    - Point plot has categorical variable on x-axis
    
- Point plots vs. bar plots
- Both show:
    - Mean of quantitative variable
    - 95% condence intervals for the mean    

```python
# Displaying the median
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import median

sns.catplot(x="smoker",y="total_bill",data=tips,kind="point",estimator=median)
plt.show()
```

```python
# Customizing the confidence intervals
import matplotlib.pyplot as plt
import seaborn as sns
sns.catplot(x="smoker",y="total_bill",data=tips,kind="point",capsize=0.2)
plt.show()
```

```python
# Turning off confidence intervals
import matplotlib.pyplot as plt
import seaborn as sns
sns.catplot(x="smoker",y="total_bill",data=tips,kind="point",ci=None)
plt.show()
```

```python
# Create a point plot of family relationship vs. absences
sns.catplot(x="famrel",y="absences",data=student_data,kind="point")
            
# Show plot
plt.show()
```
